```python
# Importing Relevant Dependencies
import pandas as pd
from matplotlib import pyplot as plt
import scipy.stats as st
import datetime as dt

filePath = "../Resources/ethVsHosp.csv"
```

```python
ethDf = pd.read_csv(filePath)

ethDf.head()
```

```python
ethDf['Race/Ethnicity'].value_counts()
```

```python
#Finding the number of hospitalized for each ethnicity

yesWsize = len(ethDf.loc[(ethDf['Race/Ethnicity'] == 'White, Non-Hispanic') & (ethDf['Hospitalized'] == 'Yes')])
yesHlSize = len(ethDf.loc[(ethDf['Race/Ethnicity'] == 'Hispanic/Latino') & (ethDf['Hospitalized'] == 'Yes')])             
yesBSize = len(ethDf.loc[(ethDf['Race/Ethnicity'] == 'Black, Non-Hispanic') & (ethDf['Hospitalized'] == 'Yes')])
yesMSize = len(ethDf.loc[(ethDf['Race/Ethnicity'] == 'Multiple/Other, Non-Hispanic') & (ethDf['Hospitalized'] == 'Yes')])
yesASize = len(ethDf.loc[(ethDf['Race/Ethnicity'] == 'Asian, Non-Hispanic') & (ethDf['Hospitalized'] == 'Yes')])
yesPiSize = len(ethDf.loc[(ethDf['Race/Ethnicity'] == 'Native Hawaiian/Other Pacific Islander, Non-Hispanic') & (ethDf['Hospitalized'] == 'Yes')])
yesNaSize = len(ethDf.loc[(ethDf['Race/Ethnicity'] == 'American Indian/Alaska Native, Non-Hispanic') & (ethDf['Hospitalized'] == 'Yes')])
```

```python
#Finding the number of not hospitalized for each ethnicity

noWsize = len(ethDf.loc[(ethDf['Race/Ethnicity'] == 'White, Non-Hispanic') & (ethDf['Hospitalized'] == 'No')])
noHlSize = len(ethDf.loc[(ethDf['Race/Ethnicity'] == 'Hispanic/Latino') & (ethDf['Hospitalized'] == 'No')])             
noBSize = len(ethDf.loc[(ethDf['Race/Ethnicity'] == 'Black, Non-Hispanic') & (ethDf['Hospitalized'] == 'No')])
noMSize = len(ethDf.loc[(ethDf['Race/Ethnicity'] == 'Multiple/Other, Non-Hispanic') & (ethDf['Hospitalized'] == 'No')])
noASize = len(ethDf.loc[(ethDf['Race/Ethnicity'] == 'Asian, Non-Hispanic') & (ethDf['Hospitalized'] == 'No')])
noPiSize = len(ethDf.loc[(ethDf['Race/Ethnicity'] == 'Native Hawaiian/Other Pacific Islander, Non-Hispanic') & (ethDf['Hospitalized'] == 'No')])
noNaSize = len(ethDf.loc[(ethDf['Race/Ethnicity'] == 'American Indian/Alaska Native, Non-Hispanic') & (ethDf['Hospitalized'] == 'No')])
```

```python
#Making my bar plots

index = ['White, Non-Hispanic','Hispanic/Latino', 'Black, Non-Hispanic', 'Multiple/Other, Non-Hispanic', 
         'Asian, Non-Hispanic', 'Native Hawaiian/Other Pacific Islander, Non-Hispanic','American Indian/Alaska Native, Non-Hispanic']
hospDf = pd.DataFrame({'Hospitalized': [yesWSize, yesHlSize,yesBSize,yesMSize,yesASize,yesPiSize,yesNaSize ], 
                       'Not Hospitalized': [noWsize, noHlSize,noBSize,noMSize,noASize,noPiSize,noNaSize]}, index=index)
hospDf.plot.bar(rot=0)
plt.legend(loc='center right')
plt.xticks(rotation='vertical')
plt.ylabel('Number of Confirmed Cases')
plt.title('Labratory Confirmed Cases Across the USA')

plt.savefig('../Images/ethVsHosp.png')
```

```python
yesHospDf = pd.DataFrame({'Race/Ethnicity': ['White, Non-Hispanic','Hispanic/Latino', 'Black, Non-Hispanic', 'Multiple/Other, Non-Hispanic', 
         'Asian, Non-Hispanic', 'Native Hawaiian/Other Pacific Islander, Non-Hispanic','American Indian/Alaska Native, Non-Hispanic'], 
                        'Number Hospitalized':[yesWSize, yesHlSize,yesBSize,yesMSize,yesASize,yesPiSize,yesNaSize ]})
yesHospDf.plot.bar(x='Race/Ethnicity', y='Number Hospitalized', rot=0, legend=False)
plt.ylabel('Number of Hospitalizations')
plt.xticks(rotation='vertical')
plt.title('Hospitalized Patients with COVID-19')

plt.savefig('../Images/allEthHosp.png')
```
